```python
# import pandas
import pandas as pd

# import numpy
import numpy as np

# import seaborn
import seaborn as sb

# import matplotlib
import matplotlib.pyplot as plt
```

```python
# loads wine dataset
from sklearn.datasets import load_wine

# load data 
wine_data=load_wine(return_X_y=False)

# description
print(wine_data.DESCR)
```

#STUDY DATASET
## For the next step, we have to check what technical information contained in the data,

```python
# load data 
X=wine_data.data
# load labels 
y=wine_data.target

# shape of data 
print('Data shape:',X.shape)
# features name 
print('Features names:',wine_data.feature_names)
# targets names 
print('Targets names:',wine_data.target_names)
```

```python
# make pandas dataframe 
wine_df=pd.DataFrame(data=X,columns=wine_data.feature_names)
# add labels 
wine_df['labels']=y
# see head 
wine_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#importing module
from sklearn.preprocessing import MinMaxScaler
# creating normalization object 
norm = MinMaxScaler()
# fit data
norm_fit = norm.fit(X_train)
new_xtrain = norm_fit.transform(X_train)
new_xtest = norm_fit.transform(X_test)
```

## Applying model 

This is the last step where we apply any suitable model which will give more accuracy, here we will use RandomForestClassifier because it was the only ML model that gives the 88% accuracy which was considered as the best accuracy.

```python
# importing modules
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,mean_squared_error
#creating RandomForestClassifier constructor
rnd = RandomForestClassifier()
# fit data
fit_rnd = rnd.fit(new_xtrain,y_train)
# predicting score
rnd_score = rnd.score(new_xtest,y_test)
print('score of model is : ',rnd_score)
# display error rate
print('calculating the error')
# calculating mean squared error
y_predict = rnd.predict(new_xtest)
rnd_MSE = mean_squared_error(y_test,y_predict)
# calculating root mean squared error
#rnd_RMSE = np.sqrt(MSE)
# display MSE
print('mean squared error is : ',rnd_MSE)
# display RMSE
#print('root mean squared error is : ',rnd_RMSE)
print(classification_report(y_test,y_predict))
```

```python
# make dict 
predicted_df={'predicted_values':y_predict,'original_values':y_test}
# creating new dataframe
pd.DataFrame(predicted_df).head(20)
```

```python
# save model 
import pickle
file_name='wine_quality_model'
# save file
save=pickle.dump(rnd,open(file_name,'wb'))
```
